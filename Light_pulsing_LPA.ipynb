```python
#converting index value into row/column space (A1->D6)

def rowcol(i):
    if i == 0:
        row = 0
        col = 0
    elif i ==1:
        row = 0
        col = 1
    elif i ==2:
        row = 0
        col = 2
    elif i ==3:
        row = 0
        col = 3
    elif i ==4:
        row = 0
        col = 4
    elif i ==5:
        row = 0
        col = 5
    elif i ==6:
        row = 1
        col = 0
    elif i ==7:
        row = 1
        col = 1
    elif i ==8:
        row = 1
        col = 2
    elif i ==9:
        row = 1
        col = 3
    elif i ==10:
        row = 1
        col = 4
    elif i ==11:
        row = 1
        col = 5
    elif i ==12:
        row = 2
        col = 0
    elif i ==13:
        row = 2
        col = 1
    elif i ==14:
        row = 2
        col = 2
    elif i ==15:
        row = 2
        col = 3
    elif i ==16:
        row = 2
        col = 4
    elif i ==17:
        row = 2
        col = 5
    elif i ==18:
        row = 3
        col = 0
    elif i ==19:
        row = 3
        col = 1
    elif i ==20:
        row = 3
        col = 2
    elif i ==21:
        row = 3
        col = 3
    elif i ==22:
        row = 3
        col = 4
    elif i ==23:
        row = 3
        col = 5
    return (row, col)
```

```python
#setting the "on" for the green wells to 7µmol/m2/s 
#and the constrant red light wells to 10 µmol/m2/s for the LPA Jennie, can adjust accordingly

LPA_name = 'Jennie'
[green_Jennie,red_Jennie]=LPA_gen('Jennie',[7],24,[10],24,0)
```

```python
import pandas as pd
#Generating pulse functions arrays

### ALL UNITS for function generation are in seconds
#Input delay term here for the delay to the first pulse
delay = 60*6*60 #6 hour delay, in secondss
delay = delay*1000 #converting to ms

#Importing step size 
step_size = 1#seconds units here too, 30 second step size

#import period/duty cycle excel file here
#see "dc_period_Jennie.xlsx" for an example of file structure
df_dc_per = pd.read_excel('dc_period_Jennie.xlsx')#replace file name here
Period_short = df_dc_per['Period']
duty_cycle_short = df_dc_per['dc']
```

```python
# Define the basic parameters of our device
#############
rowNum = 4
colNum = 6 # e.g. 24-well plate
ledsPerWell = 2
totalTime = 259200+(delay/1000) # # seconds -- total experment time time, addding the delay before induction
timeStep = step_size*1000

# Define time (when inputs change, in ms) and intensity (in GS units) matrices
#############
# Note: don't forget to add timeStep to the total time; total time must be in ms!
times = np.arange(0, totalTime*1000 + timeStep, timeStep)
#note e
# Note: the dtype of the intensity array is important for writing correct LPF files!
gsints = np.zeros((len(times), rowNum, colNum, ledsPerWell), dtype=np.int16)
## Structure of this matrix: [time][row][col][channel]
len_gsints=len(gsints)
#printing amount of timepoints per LED
print(len_gsints)
```

```python
#saving data into gs

p_array = Period_short
dc_array = duty_cycle_short

# Set row/column characterization matrix
rm = range(rowNum*colNum)

# Update gsints matrix:
## PWM Input wells:
for i in range(24): # all the remaining wells
    row, col = rowcol(rm[i])
    p = p_array[i]
    dc = dc_array[i]
    for ti,t in enumerate(times): # iterate through all timesteps
        if t < delay: #if time is below the delay, only have the red light on
            r = red_Jennie[i]
            g = 0
        elif ((t-delay)%p)/float(p) < dc: # fraction of the period < dc * period --> green LED high, red low
            r = red_Jennie[i]
            g = green_Jennie[i]
        else: # green low, red high
            r = red_Jennie[i]
            g = 0
        gsints[ti,row,col,1] = r
        gsints[ti,row,col,0] = g
gsints.resize(len_gsints, 48)


#save to CSV randomized period/duty cycle locations
#first create a pandas dataframe with necessary information

rand_dict = {'Period': p_array, 'Duty_cycle': dc_array}
rand_df = pd.DataFrame(data=rand_dict)

#s
rand_df.to_csv(+LPA_name+'pulse_parameters.csv')
```
